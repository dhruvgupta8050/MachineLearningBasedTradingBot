```python
def login(obj, client_id, password, totp):
    try:
        data = obj.generateSession(client_id, password, pyotp.TOTP(totp).now())
        
        # fetch Authorization token
        authorization_token = data['data']['jwtToken']
        # fetch Refresh token
        refresh_token = data['data']['refreshToken']
        # fetch the Feed token
        feed_token = obj.getfeedToken()

        #Fetch User Profile
        user_profile = obj.getProfile(refresh_token)
        print("USER PROFILE :\n~~~~~~~~~~~~~")
        print(user_profile)
        print("\nLogin successfull\n")
        
        # return tokens
        return authorization_token, refresh_token, feed_token
        
    except Exception as e:
        print("Login failed: {}".format(e.message))

def logout(obj, client_id):
    try:
        logout = obj.terminateSession(client_id)
        print("\nLogout Successfull\n")
    except Exception as e:
        print("Logout failed: {}".format(e.message))
```

```python
authorization_token, refresh_token, feed_token = login(obj, client_id, password, totp)
logout(obj, client_id)
print('Authorization Token : ',authorization_token,end='\n\n')
print('Refresh Token : ',refresh_token,end='\n\n')
print('Feed Token : ',feed_token,end='\n\n')
```

```python
def get_market_data_in_batches(obj, mode, exchange, tokens, batch_size=50):
    # Split tokens list into batches
    batches = [tokens[i:i + batch_size] for i in range(0, len(tokens), batch_size)]
    results = []

    for batch in batches:
        exchange_tokens = {exchange : batch}
        response = obj.getMarketData(mode, exchange_tokens)
        if response['status']:
            results.extend(response['data']['fetched'])
        else:
            print(f"Request failed: {response['message']}")

    return results

from datetime import datetime

def convert_date_format(date_str):
    # Parse the input date string
    date_obj = datetime.strptime(date_str, "%Y-%m-%d")
    # Format it to the desired output
    return date_obj.strftime("%d%b%y").upper()

# Example usage
date_str = "2024-01-06"
converted_date = convert_date_format(date_str)
print(converted_date)  # Output: 06NOV24


# Example usage
mode = "FULL"
exchange = "NFO"
tokens = d
market_data = get_market_data_in_batches(obj, mode, exchange, tokens)
data = []
for i in market_data:
    best_bid = max(i['depth']['buy'], key=lambda x: x['price'])['price']
    best_ask = min(i['depth']['sell'], key=lambda x: x['price'])['price']
    data.append([i['tradingSymbol'],best_bid,best_ask])
sorted_data = sorted(data, key=lambda x: x[0])
sorted_data
```

```python
def function_1(instrument_name : str, expiry_date : str, side : str) -> pd.DataFrame:
    
    expiry_date = convert_date_format(expiry_date)
    scrip = instrument_name + expiry_date
    
    scrip_data = obj.searchScrip("NFO", scrip)
    tokens = [item['symboltoken'] for item in scrip_data['data'] if item['tradingsymbol'].endswith(side)]
    
    market_data = get_market_data_in_batches(obj, "FULL", "NFO", tokens)
    
    data = []
    for i in market_data:
        
        token = i['symbolToken']
        option_str = i['tradingSymbol']
        first_digit_index = next(i for i, c in enumerate(option_str) if c.isdigit())
        symbol = option_str[:first_digit_index]
        strike_price_start = first_digit_index + 7  # Skip date part (2 digits + 3 letters + 2 digits)
        strike_price = option_str[strike_price_start:-2]
        option_type = option_str[-2:]
        
        best_bid = max(i['depth']['buy'], key=lambda x: x['price'])['price']
        best_ask = min(i['depth']['sell'], key=lambda x: x['price'])['price']
        
        data.append([token,symbol,strike_price,option_type,best_bid,best_ask])
            
    data = sorted(data, key=lambda x: x[2])
    data = pd.DataFrame(data,columns=['Token','Instrument','Strike Price','Side','Best Bid Price','Best Ask Price'])
    
    if side == "CE":
        view = data.drop(columns=['Token','Best Bid Price'])
    elif side == "PE":
        view = data.drop(columns=['Token','Best Ask Price'])
    return data, view
```
